```python
import cv2 as cv
```

```python
while True:
    path = input("Enter the path of your image: ")
    print("The path of your image is:", path)

    typ = int(input("Which type of image is it (if color => 1; else => 0): "))

    if typ == 1:
        x = "color image"
    elif typ == 0:
        x = "black and white / grey image"
    else:
        print("Give 0 or 1 only")
        continue

    print(f"You have chosen: {typ}. This means your image is: {x}")

    or_img = cv.imread(path, typ)

    while True:
        choice = int(input("Enter to convert your image.\n1 => Color image.\n0 => Black and white/grey image\n2 => Exit: "))
        print("You entered:", choice)

        if (choice == 1 and typ == 1) or (choice == 0 and typ == 0):
            cvt_img = cv.imread(path, -1)
            if or_img.shape[1] >= or_img.shape[0] + 200:
                or_img = cv.resize(or_img, (1500,750))
            elif or_img.shape[1] + 200 <= or_img.shape[0]:
                or_img = cv.resize(or_img,(700,775))
            else:
                or_img = cv.resize(or_img,(750,750))
            cvt_img = cv.resize(cvt_img, (or_img.shape[1],or_img.shape[0]))

            cv.imshow("Original Image", or_img)
            cv.imshow("Converted Image", cvt_img)

            k = cv.waitKey(0)
            print("----------------------------------------------------------------\nPress 's' to download the converted image")

            if k == ord('s'):
                cvt_name = input("Give the name of your converted image: ")
                print("The name you gave for your converted image is", cvt_name)
                cv.imwrite(f"C:\\Users\\Hp\\Downloads\\{cvt_name}.png", cvt_img)
                cv.destroyAllWindows()
            else:
                cv.destroyAllWindows()
        elif choice ==1 or choice ==0:
            cvt_img = cv.imread(path,choice)
            if or_img.shape[1] >= or_img.shape[0] + 200:
                or_img = cv.resize(or_img, (1500,750))
            elif or_img.shape[1] + 200 <= or_img.shape[0]:
                or_img = cv.resize(or_img,(700,775))
            else:
                or_img = cv.resize(or_img,(750,750))
            cvt_img = cv.resize(cvt_img, (or_img.shape[1],or_img.shape[0]))

            cv.imshow("Original Image", or_img)
            cv.imshow("Converted Image", cvt_img)

            k = cv.waitKey(0)
            print("----------------------------------------------------------------\nPress 's' to download the converted image")

            if k == ord('s'):
                cvt_name = input("Give the name of your converted image: ")
                print("The name you gave for your converted image is", cvt_name)
                cv.imwrite(f"C:\\Users\\Hp\\Downloads\\{cvt_name}.png", cvt_img)
                cv.destroyAllWindows()
            else:
                cv.destroyAllWindows()

        elif choice == 2:
            break
        else:
            print("Please enter a valid choice (0, 1, or 2)")

    again = int(input("Enter:\n1 => Continue to enter another path of the image\n0 => Stop/Exit.\n"))
    print(f"You have entered: {again}")

    if again == 0:
        print("----------------------------------------------------------------\nThank you.\nHope you enjoyed our service.\nVisit again.")
        break
    if again !=1 and again !=0:
        print("Enter the valid answer 0/1",again)
```
